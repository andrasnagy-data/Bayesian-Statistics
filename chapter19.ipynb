```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
hypotheses = np.linspace(0, 1, 1000)
```

```python
def bayes_factor(h1, h2):
    return (h1**24) * ((1-h1)**76) / (h2**24) * ((1-h2)**76)
```

```python
bayes_factor1 = bayes_factor(hypotheses, 0.5)
bayes_factor2 = bayes_factor(hypotheses, 0.24)
```

```python
plt.plot(hypotheses, bayes_factor1);
```

```python
plt.plot(hypotheses, bayes_factor2);
```

```python
fig, ax = plt.subplots()
ax.plot(hypotheses, bayes_factor1/sum(bayes_factor1), 'g')
ax.plot(hypotheses, bayes_factor2/sum(bayes_factor2), 'y'); # color of plot -> two plots are identical
```
