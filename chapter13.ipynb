```python
from scipy.stats import beta, norm
import numpy as np
import matplotlib.pyplot as plt
```

#### Exercise 1.

```python
a, b = 300, 40000-300
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
plt.figure(figsize= (12, 8))
plt.title('PDF of Beta(a, b)')
plt.xlabel('Density')
plt.ylabel('Probability of Subscription')
plt.plot(x, beta.pdf(x, a, b));
```

```python
plt.figure(figsize= (12, 8))
plt.title('CDF of Beta(a, b)')
plt.xlabel('Probability of Subscription')
plt.ylabel('Cumulative probability')
plt.plot(x, beta.cdf(x, a, b));
```

#### Exercise 2.

```python
data = np.array([7.8, 9.4, 10, 7.9, 9.4, 7, 7, 7.1, 8.9, 7.4])
mean = np.mean(data)
std = np.sqrt(np.var(data))

lower = norm.ppf(0.0005, mean, std)
upper = norm.ppf(0.9995, mean, std)

f"The 99.9% confidence interval of true value of snowfall is: {lower:.3f} cm and {upper:.3f} cm"
```
