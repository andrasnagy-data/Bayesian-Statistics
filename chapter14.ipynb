```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

#### Exercise 1.

```python
a, b = 7, 3
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)

plt.xlabel('True probability of flipping heads')
plt.ylabel('density')
plt.plot(x, beta.pdf(x, a, b));
```

#### Exercise 2.

```python
lower = beta.ppf(0.025, 7+9, 3+11)
upper = beta.ppf(0.975, 7+9, 3+11)

f"The 95% confidence interval for true rate of heads is between: {lower:.4f} and {upper:.4f}"
```
